# Gradient Descent (In general)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X - 5)**2

print(X, Y)
```

**Goal**: Given a function $f(x)$, find the value of $x$ that minimizes $f$

## Visualization

```python
plt.style.use("seaborn")

plt.title("Y = (X-5)^2")
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X") 
plt.show()
```

## Gradient Descent Algorithm
- Step 1 : Initialize random x value
- Step 2 : Move Downhill by gradually updating the value of $x$ using formula: $x = x - \eta \frac{\partial f(x)}{\partial x}$
 
- Step 3 : Stop when $\frac{\partial f(x)}{\partial x}$ approxes to or equals to 0, that is change in $x$ is zero or very small.

```python
x = 0
lr = 0.1
error = []
plt.plot(X,Y)

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    Y = (x-5)**2
    error.append(Y)
    plt.scatter(x, Y)
    print(x)
```
